```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def simple_plot_2_arrays(x_axis_array, y_axis_array, label_dict={}, save_as=None):
    """
    label_dict should have 'xlabel', 'ylabel', 'title'.
    """
    fig, ax = plt.subplots()
    ax.plot(x_axis_array, y_axis_array)

    ax.set(**label_dict)
    ax.grid()

    if save_as:
        fig.savefig(save_as)
    plt.show()

def simple_plot_x2y_dict(x2y_dict, label_dict={}, save_as=None):
    x_axis_array, y_axis_array = zip(*x2y_dict.items())
    simple_plot_2_arrays(x_axis_array, y_axis_array, label_dict=label_dict, save_as=save_as)
```

```python
d = {2000: 0.003942828979793003, 4000: 0.26231272988247956, 6000: 0.3390721275824574, 8000: 0.3773441939761319}
simple_plot_x2y_dict(d, label_dict={"xlabel":"#-training-samples", "ylabel":"Arg-F1", "title":"QANom Learning Curve"}, save_as="plots/learning_curve_qanom.png")
```
